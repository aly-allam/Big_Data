```python
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
from pylab import rcParams
sns.set_style('darkgrid')
from skimpy import clean_columns

import warnings
warnings.filterwarnings('ignore')
```

```python
import plotly 
import plotly.express as px
import plotly.graph_objs as go
import plotly.offline as py
from plotly.offline import iplot
from plotly.subplots import make_subplots
import plotly.figure_factory as ff
```

```python
df_2017 = pd.read_csv('2017.csv',index_col=False)
df_2018 = pd.read_csv('2018.csv',index_col=False)
df_2019 = pd.read_csv('2019.csv',index_col=False)
df_2020 = pd.read_csv('2020.csv',index_col=False)
```

```python
goals_2017 = df_2017['Goals'].head(2)
comp_2017 = df_2017['Competition'].head(2)
mpg_2017 = df_2017['Minutes Per Goal'].head(2)

fig = plt.figure(figsize =(5, 5))
plt.bar(comp_2017, goals_2017,label="Goals")
plt.xlabel("Competition")
plt.ylabel("Goals")
plt.title("Mohamed Salah goals with Liverpool FC in Premier League and Champions League 2017/2018 season")
plt.ylim(0,40)
plt.legend()
plt.show()
```

```python
mpg_2017
```

```python
for x in range(len(mpg_2017)):
    mpg_2017[x] = mpg_2017[x][0:-1]
```

```python
mpg_2017
```

```python
fig = plt.figure(figsize =(5, 5))
plt.bar(comp_2017, [int(x) for x in mpg_2017],label="Minutes Per Goal", color = "red")
plt.xlabel("Competition")
plt.ylabel("Minutes Per Goal")
plt.title("Mohamed Salah minutes per goal with Liverpool FC in Premier League and Champions League 2017/2018 season")
plt.ylim(85,100)
plt.legend()
plt.show()
```
